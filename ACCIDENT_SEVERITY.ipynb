Since there are so many categorical values, we need to use feature selection We need to perform label encoding before applying chi 2 analysis

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

#creating a new data frame from performing the chi2 analysis
df1=pd.DataFrame()

#adding all the categorical columns except the output to new data frame
for i in categorical:
    if i!= 'Accident_severity':
        df1[i]=le.fit_transform(df[i])
```

```python
df1.info()
```

```python
df1.corr()
```

```python
plt.figure(figsize=(22,17))
sns.set(font_scale=1)
sns.heatmap(df1.corr(), annot=True)
```

```python
from sklearn.feature_selection import chi2
f_p_values=chi2(df1,df['Accident_severity'])
```

```python
f_p_values
```

```python
f_p_values1=pd.DataFrame({'features':df1.columns, 'Fscore': f_p_values[0], 'Pvalues':f_p_values[1]})
f_p_values1
```

```python
#since we want lower Pvalues we are sorting the features
f_p_values1.sort_values(by='Pvalues',ascending=True)
```

we need higher Fscore and lower the Pvalues, so by evaluating, we can remove Owner_of_vehicle, Type_of_vehicle, Road_surface_conditions, Pedestrian_movement,Casualty_severity,Educational_level,Day_of_week,Sex_of_driver,Road_allignment, Sex_of_casualty

after evaluating we are removing lesser important columns and storing to a new data frame

```python
df2=df.drop(['Owner_of_vehicle', 'Type_of_vehicle', 'Road_surface_conditions', 'Pedestrian_movement',
         'Casualty_severity','Educational_level','Day_of_week','Sex_of_driver','Road_allignment',
         'Sex_of_casualty'],axis=1)
df2.head()
```

```python
df2.shape
```

```python
df2.info()
```

to check distinct values in each categorical columns we are storing them to a new variable

```python
categorical_new=[i for i in df2.columns if df2[i].dtype=='O']
print(categorical_new)
```

```python
dummy = pd.get_dummies(df2[['Age_band_of_driver', 'Vehicle_driver_relation', 'Driving_experience', 'Area_accident_occured', 'Lanes_or_Medians', 'Types_of_Junction', 'Road_surface_type', 'Light_conditions', 'Weather_conditions', 'Type_of_collision', 'Vehicle_movement', 'Casualty_class', 'Age_band_of_casualty', 'Cause_of_accident']])
dummy.head()
```

concatinate dummy and old data frame

```python
df3=pd.concat([df2,dummy],axis=1)
df3.head()
```

dropping the dummy variable

```python
df3.drop(['Age_band_of_driver', 'Vehicle_driver_relation', 'Driving_experience', 'Area_accident_occured', 'Lanes_or_Medians','Types_of_Junction', 'Road_surface_type', 'Light_conditions', 'Weather_conditions', 'Type_of_collision','Vehicle_movement','Casualty_class', 'Age_band_of_casualty', 'Cause_of_accident'],axis=1,inplace=True)
df3.head()
```

```python
x=df3.drop(['Accident_severity'],axis=1)
x.shape
```

```python
x.head()
```

```python
y=df3.iloc[:,2]
y.head()
```

```python
y.value_counts()
```

```python
sns.countplot(x=y)
```

DATA IMBALANCE:


```python
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
xo,yo=oversample.fit_resample(x,y)
```

checking the oversampling output

```python
y1=pd.DataFrame(yo)
y1.value_counts()
```

```python
sns.countplot(x = yo, palette='muted')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(xo,yo,test_size=0.3,random_state=42)
```

```python
pd.set_option('display.max_columns',None)
x_test
```

```python
x_train
```

```python
y_test
```

KNN


```python
y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier(n_neighbors=5)
KNN.fit(x_train,y_train)
```

```python
y_KNN = KNN.predict(x_test)
y_KNN
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,ConfusionMatrixDisplay
matrix_KNN=confusion_matrix(y_test,y_KNN)
print(matrix_KNN,'\n')
print(ConfusionMatrixDisplay.from_predictions(y_test,y_KNN))
accuracy_KNN=accuracy_score(y_test,y_KNN)
print(accuracy_KNN,'\n')
report_KNN=classification_report(y_test,y_KNN)
print(report_KNN)
```

```python
KNN.score(x_test,y_test)
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_LR=lr.predict(x_test)
y_LR
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,ConfusionMatrixDisplay
matrix_LR=confusion_matrix(y_test,y_LR)
print(matrix_LR,'\n')
print(ConfusionMatrixDisplay.from_predictions(y_test,y_LR))
accuracy_LR=accuracy_score(y_test,y_LR)
print(accuracy_LR,'\n')
report_LR=classification_report(y_test,y_LR)
print(report_LR)
```

```python
lr.score(x_test,y_test)
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier(criterion='entropy')
DT.fit(x_train,y_train)
```

```python
y_DT=DT.predict(x_test)
y_DT
```

```python
matrix_DT=confusion_matrix(y_test,y_DT)
print(matrix_DT,'\n')
print(ConfusionMatrixDisplay.from_predictions(y_test,y_DT))
accuracy_DT=accuracy_score(y_test,y_DT)
print(accuracy_DT,'\n')
report_DT=classification_report(y_test,y_DT)
print(report_DT)
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=10,criterion='entropy')
rf.fit(x_train,y_train)
```

```python
y_rf = rf.predict(x_test)
y_rf
```

```python
rf.score(x_test,y_test)
```
